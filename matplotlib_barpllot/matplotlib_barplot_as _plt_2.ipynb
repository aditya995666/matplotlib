# bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# dataset of indian ai intelligent
classes = ['AI',"ML",'DS',"PY",'R']
class_1=[23,24,25,26,27]
class_2=[24,25,26,27,28]
class_3=[25,26,27,28,29]
```

```python
plt.bar(classes,class_1)
plt.show()
```

```python
plt.bar(classes,class_1 ,width=0.2)
plt.show()
```

```python
plt.bar(classes,class_1 ,width=0.2,align="center")
plt.show()
```

```python
plt.bar(classes,class_1 ,width=0.2,align="center",color="r",edgecolor="g",linewidth=2)
plt.show()
```

```python
plt.bar(classes,class_1 ,width=0.2,align="center",color="r",edgecolor="g",linewidth=2,linestyle="--")
plt.show()
```

```python
plt.bar(classes,class_1 ,width=0.2,label="class_1 student",align="center",color="r",edgecolor="g",linewidth=2,linestyle="--")
plt.legend()
plt.show()
```

# multiple bar chart

```python
from matplotlib import style
style.use("ggplot")
plt.bar(classes,class_1 ,width=0.2,label="class_1 student",align="center",color="r",edgecolor="g",linewidth=2,linestyle="--")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
style.use("ggplot")
plt.bar(classes,class_1 ,width=0.2,label="class_1 student",align="center",color="black",edgecolor="g",linewidth=2,linestyle="--")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
style.use("ggplot")
plt.bar(classes,class_1 ,width=0.2,label="class_1 student",align="center",color="black",edgecolor="g",linewidth=2,linestyle="--")
plt.legend()
plt.title("AI vs ML",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("students",fontsize=15)
plt.show()
```

```python
from pathlib import WindowsPath
plt.figure(figsize=(10,5))
style.use("ggplot")
width=0.2
classes_index=np.arange(len(classes))
plt.bar(classes_index,class_2 ,width,label="class_2 student",align="center",color="red",edgecolor="g",linewidth=2,linestyle="--")

plt.bar(classes_index + width ,class_1 ,width,label="class_1 student",align="center",color="black",edgecolor="g",linewidth=2,linestyle="--")
plt.legend()
plt.title("AI vs ML",fontsize=15)
plt.xlabel("classes",fontsize=15)
plt.ylabel("students",fontsize=15)
plt.show()
```
