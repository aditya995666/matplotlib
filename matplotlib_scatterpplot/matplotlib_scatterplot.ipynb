# scatter_plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/googleplaystore.csv",nrows=500)
df.head()
```

```python
x=df["Rating"]
y=df["Reviews"]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Rating vs Reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.scatter(x,y ,color="r",linewidth=0.2  )
plt.title("Rating vs Reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.scatter(x,y ,color="r",marker="*" ,linewidths=10)
plt.figure(figsize=(10,5))
plt.title("Rating vs Reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.scatter(x,y ,color="r",edgecolors="g"  ,marker="*" ,linewidths=2)
plt.figure(figsize=(10,5))
plt.title("Rating vs Reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.scatter(x,y ,color="r",edgecolors="g"  ,marker="*" ,linewidths=2)

plt.scatter(x,df["Installs"] ,color="b",edgecolors="g"  ,marker="o" ,linewidths=2)

plt.figure(figsize=(10,5))
plt.title("Rating vs Reviews")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```
