# plot *pie_chart*

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
classess=["python","java", "ml","ai","r"]
class_student=[23,56,45,28,90]
```

```python
plt.pie(class_student,labels=classess)
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0,0,0.2,0,0])
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0.3,0.2,0.2,0,0])
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0.3,0.2,0.2,0,0],colors=["y","r","b","g","c"])
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0.3,0.2,0.2,0,0],colors=["y","r","b","g","c"],
        autopct="%0.1f%%")
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0.3,0.2,0.2,0,0],colors=["y","r","b","g","c"],
        autopct="%0.1f%%",
        shadow=True)
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0,0,0.2,0,0],colors=["y","r","b","g","c"],
        autopct="%0.1f%%",
        shadow=True,
        startangle=90) # angle
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0,0,0.2,0,0],colors=["y","r","b","g","c"],
        autopct="%0.1f%%",
        shadow=True,
        startangle=90,
        radius=3,textprops={"fontsize":15}) # angle
plt.show()
```

```python
plt.pie(class_student,labels=classess   ,explode=[0,0,0.2,0,0],colors=["y","r","b","g","c"],
        autopct="%0.1f%%",
        shadow=True,
        startangle=90,
        radius=3,textprops={"fontsize":15}) # angle
plt.legend(title="class")
plt.show()
```
